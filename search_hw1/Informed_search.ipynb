```python
import time
import random
def read_words_file():
    """
    read words from dictionary,
    return words list
    """
    with open ("Words.txt", "r") as myfile:
        dicts=myfile.read()
        dicts = dicts.split()
    return dicts
```

```python
def random_pick(dicts):
    """
    randomly pick two words with same 
    length in the dicts
    """
    start=random.choice(dicts)
    random.shuffle(dicts)
    for i in dicts:
        if len(i)== len(start):
            end=i
            return start, end    
```

```python
def find_path(startWord,visited,parent,string):
    """
    after find the endword, use this function 
    trace back through the parent attribute to 
    the startword. Eventually return a path list
    contain all the word in the path.
    """
    path=[string]
    for ind in range(len(visited)):
            if parent == visited[ind].string:
                path.insert(0,visited[ind].string)
    while True:
        for ind2 in range(len(visited)):
            if path[0] == visited[ind2].string:
                path.insert(0,visited[ind2].parent)
                continue

            if path[0]== startWord:
                return path
```

```python
def greedy(word,endWord,steps):
    """
    get h(x) points how many characters are wrong
    get g(x) indicate how many steps each node have
    return g(x)+h(x)
    """
    score=0
    for ind in range(len(endWord)):
        temp = word[ind]
        #print(temp)
        if temp != endWord[ind]:
            score = score+1
            
    return score+steps
```

```python
class word_obj(object):
    '''
    create word object to contain score, string 
    and parent
    '''
    def __init__(self, string, score=0, parent=""):
        self.string = string
        self.score = score
        self.parent = parent
```

```python
def find_neighbours(word,wordList):
    """
    check letters in word from a to z 
    generate different words which only one
    letter different from orignal words
    compare with dicts and word itsself,
    if we can find change in dicts then
    append the result to a list6
    """
    word=''.join(word)
    neighbours= list()    
    for ind in range(len(word)):
        temp = word[ind]
        for alphbet in [chr(x) for x in range(ord('a'), ord('z')+1)]:
            if alphbet != temp:
                new_word = word[:ind]+ alphbet +word[ind+1:]
                if new_word != word and new_word in wordList:
                    neighbours.append(new_word)
                    wordList.remove(new_word)
                    
                    
    #print("word:",word,"neighbours:",neighbours)                
    return neighbours
```

```python
def informed_search(startWord, endWord, wordList):
    """
    Try to use A* search and create f(x) = g(x) + h(x)
    g(x) means how many steps already have
    h(x) means how many characters are wrong
    score to search.
    """
    steps=0
    fringe=[]
    visited=[]
    wordList.remove(startWord)        
    fringe.append(word_obj(startWord,greedy(startWord,endWord,steps)))
    steps=steps+1
    x=[]
    for i in find_neighbours(fringe[0].string,wordList):
        x.append(word_obj(i,greedy(i,endWord,steps),startWord))
    fringe=x+fringe
    fringe.sort(key=lambda x: (x.score))
    while len(fringe)>=0:
        for obj in fringe:
            temp=fringe.pop(0)
            visited.append(temp)
            temp1=find_neighbours(temp.string,wordList)
            steps=steps+1
            temp3=[]
            for word in temp1:                
                temp2=word_obj(word,greedy(word,endWord,steps),temp.string)
                temp3.append(temp2)            

            fringe=temp3+fringe
            fringe=sorted(fringe,
                        key=lambda x: (x.score))
            #for i in fringe:
                #print(i.string)            
            
            for ind in range(len(fringe)):
                if endWord == fringe[ind].string:                    
                    return print("steps: ", steps,"path:", find_path(startWord,visited,fringe[ind].parent,fringe[ind].string) ) 
        if len(fringe)==0:
                return print("no solution!")  
```

```python
import numpy as np
def random100():
    """
    100 times randomly selection two word from
    dictionary and return average time.
    """    
    time1=0
    for i in range(100):
        dicts= read_words_file()
        beginWord, endWord = random_pick(dicts)
        print("begin word:", beginWord, "end word:",endWord)
        starttime=time.time()
        print(informed_search(beginWord, endWord, dicts))
        endtime=time.time()
        time_taken = endtime - starttime
        time1=time1+time_taken
    avg_time=time1/100
    return print("The 100 random informed search average time is  ",avg_time)


random100()    
```
