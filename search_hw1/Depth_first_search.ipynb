```python
import time
import random
def read_words_file():
    """
    read words from dictionary,
    return words list
    """
    with open ("Words.txt", "r") as myfile:
        dicts=myfile.read()
        dicts = dicts.split()
    return dicts
```

```python
def random_pick(dicts):
    """
    randomly pick two words with same 
    length in the dicts
    """
    start=random.choice(dicts)
    random.shuffle(dicts)
    for i in dicts:
        if len(i)== len(start):
            end=i
            return start, end    
```

```python
def find_path(startWord,visited,parent,string):
    """
    after find the endword, use this function 
    trace back through the parent attribute to 
    the startword. Eventually return a path list
    contain all the word in the path.
    """
    path=[string]
    for ind in range(len(visited)):
            if parent == visited[ind].string:
                path.insert(0,visited[ind].string)
    while True:
        for ind2 in range(len(visited)):
            if path[0] == visited[ind2].string:
                path.insert(0,visited[ind2].parent)
                continue

            if path[0]== startWord:
                return path
```

```python
class word_obj(object):
    '''
    create word object to contain score, string 
    and parent
    '''
    def __init__(self, string, score=0, parent=""):
        self.string = string
        self.score = score
        self.parent = parent
```

```python
def find_neighbours(word,wordList):
    """
    check letters in word from a to z 
    generate different words which only one
    letter different from orignal words
    compare with dicts and word itsself,
    if we can find change in dicts then
    append the result to a list6
    """
    word=''.join(word)
    neighbours= list()    
    for ind in range(len(word)):
        temp = word[ind]
        for alphbet in [chr(x) for x in range(ord('a'), ord('z')+1)]:
            if alphbet != temp:
                new_word = word[:ind]+ alphbet +word[ind+1:]
                if new_word != word and new_word in wordList:
                    neighbours.append(new_word)
                    wordList.remove(new_word)
                    
                    
    #print("word:",word,"neighbours:",neighbours)                
    return neighbours
```

```python
def dfs(startWord, endWord, wordList):
    '''
    depth first search
    use stack to search
    '''
    result = 0
    visited =[]
    temp=[]
    fringe = [word_obj(startWord)]
    wordList.remove(startWord)
    x=[]
    for i in find_neighbours(startWord,wordList):
        x.append(word_obj(i,0,startWord))
    fringe=x+fringe
    while len(fringe)>=0:
        for obj in fringe:
            temp=fringe.pop(0)
            visited.append(temp)
            result +=1
            temp1=find_neighbours(temp.string,wordList)
            temp3=[]
            for word in temp1:
                temp2=word_obj(word,0,temp.string) 
                temp3.append(temp2)
            fringe=temp3+fringe
            for ind in range(len(fringe)):
                if endWord == fringe[ind].string:
                    return print("steps: ", result,"path:", find_path(startWord,visited,fringe[ind].parent,fringe[ind].string) )
                
        if len(fringe)==0:
            return print("no solution!")  
```

```python
import numpy as np
def random100():
    """
    100 times randomly selection two word from
    dictionary and return average time.
    """    
    time1=0
    for i in range(100):
        dicts= read_words_file()
        beginWord, endWord = random_pick(dicts)
        print("begin word:", beginWord, "end word:",endWord)
        starttime=time.time()
        print(dfs(beginWord, endWord, dicts))
        endtime=time.time()
        time_taken = endtime - starttime
        time1=time1+time_taken
    avg_time=time1/100
    return print("The 100 random depth first search average time is  ",avg_time)


random100()    
```
